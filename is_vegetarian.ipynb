```python
for index, row in df_train.iterrows():
        str = row['ingredients']
        str2 = str.replace("[","").replace("]","").replace("'", "").replace(",", "")
        list = str2.split()
        list2 = row['name'].split()
        
        
        filtered_list = []
        filtered_list2 = []
        for word in list:
                if word not in remove_words:
                        filtered_list.append(word)
            # Ajoutez le mot à la nouvelle liste s'il n'est pas dans remove_words
        sequence_texte = ", ".join(filtered_list)
        for word in list2:
                if word not in remove_words:
                        filtered_list2.append(word)
            # Ajoutez le mot à la nouvelle liste s'il n'est pas dans remove_words
        sequence_texte2 = ", ".join(filtered_list2)
    
       
        
        df_train.at[index, 'ingredients'] = sequence_texte
        df_train.at[index, 'name'] = sequence_texte2
```

```python
compteur = 0
for index, row in df_train.iterrows():
    phrase = row['ingredients']
    if isinstance(phrase, float):
        phrase = "suuu" 
    phrase_embedding = model.encode(phrase, convert_to_tensor=True)
    phrase_embedding_list = phrase_embedding.cpu().numpy()
    df.at[index, 'ingredients'] = phrase_embedding_list
    phrase2 = row['name']
    if isinstance(phrase2, float):
        phrase2 = "suuu" 
    phrase_embedding2 = model.encode(phrase2, convert_to_tensor=True)
    phrase_embedding_list2 = phrase_embedding2.cpu().numpy()
    df.at[index, 'name'] = phrase_embedding_list2
    compteur = compteur + 1
    print(f'{compteur}/68191')
```

```python
y = df['tags'].astype(float)
X = df[['ingredients', 'name']]
```

```python
ingredients_df = pd.DataFrame(X['ingredients'].tolist(), columns=[f'ingredient_feature_{i}' for i in range(len(X['ingredients'].iloc[0]))])
#X_processed = ingredients_df
# Ajoutez les caractéristiques de chaque nom
# Les colonnes seront les caractéristiques des noms, nommées "name_feature_0", "name_feature_1", etc.
name_df = pd.DataFrame(X['name'].tolist(), columns=[f'name_feature_{i}' for i in range(len(X['name'].iloc[0]))])

# Concaténez les deux DataFrames pour obtenir un DataFrame contenant toutes les caractéristiques
X_processed = pd.concat([ingredients_df, name_df], axis=1)
```

```python
print(X_processed)
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
X_resampled, y_resampled = rus.fit_resample(X_processed, y)
```

```python
print(X_resampled)
```

```python
print(y_resampled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
gnb = GaussianNB()
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=100, random_state=0)
neigh = KNeighborsClassifier(n_neighbors=50)
```

```python
y_pred = gnb.fit(X_train, y_train)
```

```python
y_pred_test = gnb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_test)
print("Accuracy sur l'ensemble de test :", accuracy)
```

```python
y_pred = neigh.fit(X_train, y_train)
y_pred_test = neigh.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_test)
```
